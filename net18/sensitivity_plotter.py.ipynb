```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

scenario = 1
nome_scheda = 'Sensitivity'
df = pd.read_excel(f'./validation_data/Scenario{scenario}.xlsx', sheet_name=nome_scheda)

valori = df.values

data = abs(np.array(valori)[:18, 1:])

models_input_indices = list(range(5,22)) + list(range(22, 39)) + list(range(5)) + list(range(39, 46)) + list(range(46, 53))
data = data[:, models_input_indices]

node_index = 3
x_positions = np.arange(len(data[node_index]))

plt.figure(figsize=(15,6))
plt.bar(x_positions, data[node_index], color='green')
plt.xlabel('Measure index')
plt.ylabel('Sensitivity values')
plt.title(f'Sensitivity for node: {node_index}, scenario: {scenario}')
```
